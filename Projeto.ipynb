#### Estatísticas descritivas Utilidades domésticas e PCs


```python
dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price'].min()
```

```python
valor = dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price']
def estatistica_utd() : 
    count, mean, std, min,Q1, Q2, Q3,max = valor.describe()
    amplitude = dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price'].max() - dfProductsMerge.query('product_category_name == "utilidades_domesticas"')['price'].min()
    cv = (std/mean) * 100
    dados = {'Q1': Q1, 'Q2': Q2, 'Q3': Q3, 'Valor máximo': max,
    'Valor mínimo': min, 'Média': mean, 'Desvio padrão': std, 'Amplitude': amplitude, 
    'Coeficiente de variação': cv}
    dfEstUtd = pd.DataFrame(dados, index = ['Estatísticas Utilidades Domésticas'])
    return dfEstUtd
    
```

```python
estatistica_utd()
```

Estatísticas descritivas Agro indústria.

```python
def estatistica_pcs() : 
    valor = dfProductsMerge.query('product_category_name == "pcs"')['price']
    count, mean, std, min,Q1, Q2, Q3,max = valor.describe()
    amplitude = dfProductsMerge.query('product_category_name == "pcs"')['price'].max() - dfProductsMerge.query('product_category_name == "pcs"')['price'].min()
    cv = (std/mean) * 100
    dados = {'Q1': Q1, 'Q2': Q2, 'Q3': Q3, 'Valor máximo': max,
    'Valor mínimo': min, 'Média': mean, 'Desvio padrão': std, 'Amplitude': amplitude, 
    'Coeficiente de variação': cv}
    dfEstPcs = pd.DataFrame(dados, index = ['Estatísticas PCS'])
    return dfEstPcs
```

```python
estatistica_pcs()
```
